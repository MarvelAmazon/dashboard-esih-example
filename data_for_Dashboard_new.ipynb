```python
def upgrade(df):
    df['passed'] = 0
    for index in range(df.shape[0]):
        if (df.at[index,'grade']>= 65) and (df.at[index,'id_level'] ==1):
            df.at[index,'passed']  = 1
        elif (df.at[index,'grade']>= 70) and (df.at[index,'id_level'] ==2):
            df.at[index,'passed']  = 1
        elif (df.at[index,'grade']>= 75) and (df.at[index,'id_level'] ==3):
            df.at[index,'passed']  = 1
        elif (df.at[index,'grade']>= 80) and (df.at[index,'id_level'] ==4):
            df.at[index,'passed']  = 1
        
    return df
    
```

```python
def new_enrolment1(began_year=1996,end_year=2020,id_levels=[1,2,3,4]):
    i=0
    student_id =[]
    years = []
    levels = []
    sex = []
    vaccation = []
    level_name = []
    c =0
    lenght =0
    for year in range(began_year,end_year):
        if year == began_year:
            for id_level in id_levels:
                
                if id_level == 1:
                    lenght =100
                    lv_name = "L0"
                elif id_level==2:
                    lenght =85
                    lv_name = "L1"
                elif id_level==3:
                    lenght = 75
                    lv_name = "L2"
                elif id_level==4:
                    lenght = 60
                    lv_name = "L3"
                for index in range(0,lenght):
                    
                   
                    student_id.append(str(uuid.uuid1()))
                    sex.append(np.random.choice([0,1], p = [0.3,0.7] ))
                    years.append(year)
                    levels.append(id_level)
                    level_name.append(lv_name)
                    
        else:
            for id_level in [1,2]:
                if id_level == 1:
                    #print(year,id_level)
                    lenght =100
                    lv_name = "L0"
                elif id_level==2:
                    lenght =25
                    lv_name = "L1"
                else:
                    lenght =0
                
                for index in range(0,lenght):
                    student_id.append(str(uuid.uuid1()))
                    sex.append(np.random.choice([0,1], p = [0.3,0.7] ))
                    years.append(year)
                    levels.append(id_level)
                    level_name.append(lv_name)
   
    df_enrolement = pd.DataFrame(dict(student_id=student_id,gender=sex,year=years,id_level=levels,level_name=level_name))  
    df_enrolement['status'] = 'new enrolled'
    return df_enrolement
```

```python
def new_enrolment2(began_year=1996,end_year=2020,id_levels=[1,2,3,4]):
    i=0
    student_id =[]
    sex = []
    years = []
    levels = []
    level_name = []
    c =0
    lenght =0
    for year in range(began_year,end_year):
        
            for id_level in [1,2]:
                if id_level == 1:
                    lenght =100
                    lv_name = "L0"
                elif id_level==2:
                    lenght =15
                    lv_name = "L1"
                else:
                    lenght =0
                
                for index in range(0,lenght):
                    student_id.append(str(uuid.uuid1()))
                    sex.append(np.random.choice([0,1], p = [0.3,0.7] ))
                    years.append(year)
                    levels.append(id_level)
                    level_name.append(lv_name)
   
    df_enrolement = pd.DataFrame(dict(student_id=student_id,gender=sex,year=years,id_level=levels,level_name=level_name))  
    df_enrolement['status'] = 'new enrolled'
    return df_enrolement
```

```python
def simulate_promo(student_enrolmemnt,number_year):
    palmares=[]
    admin_list =[]
    echech_list =[]
    min_year = min(list(student_enrolmemnt.year.unique()))
    max_year = min_year+number_year+1
    for year in range(min_year,max_year):

        current_promo = student_enrolmemnt[student_enrolmemnt.year == year]
        df_check1 = pd.merge(df_ece,current_promo, how = 'inner', on='id_level')
        
        df_check1['vaccation'] = np.nan
        df_check1.loc[df_check1.id_level==1, 'vaccation'] = 0
        df_check1.loc[df_check1.id_level!=1, 'vaccation'] = np.random.binomial(1, 
                                                                     0.7, 
                                                                     len(df_check1.loc[df_check1.id_level!=1, 'vaccation']))
        df_check1['vaccation'] = df_check1['vaccation'].astype(int)
        
        x = np.random.normal(loc=75, scale=20, size=(df_check1.shape[0],))
        df_check1['grade'] =x
        df_check1.loc[df_check1.grade >100,'grade']=100
        df_check1 = upgrade(df_check1)
        df_year =df_check1.groupby(['id_level','year','student_id','gender', 'vaccation']).grade.mean().to_frame().reset_index()
        df_temp3 = upgrade(df_year)
        admins = df_temp3.loc[df_temp3.passed ==1,['student_id','year','id_level','gender', 'vaccation']]
       
        for index in admins.index:
           
            if admins.at[index,'id_level'] <=4:
                admins.at[index,'year'] =admins.at[index,'year']+1
                admins.at[index,'id_level'] =admins.at[index,'id_level']+1
                admins.at[index,'status'] ='new passed'
        echec = df_temp3.loc[df_temp3.passed ==0,['student_id','year','id_level','gender', 'vaccation']]
       
        for index in echec.index: 
            if echec.at[index,'id_level'] >=2 and echec.at[index,'id_level'] <=4:
                echec.at[index,'year'] =echec.at[index,'year']+1
                echec.at[index,'status'] ='new failed'

            else:
                echec.at[index,'status'] ='removed'
        
        echec =echec[echec.status !='removed']      
        new_student = new_enrolment2(began_year=year+1,end_year=year+2)
        student_enrolmemnt = pd.concat([student_enrolmemnt,new_student,admins,echec])
        palmares.append(df_check1)
        admin_list.append(admins)
        echech_list.append(echec)
        
    palmares =pd.concat(palmares)    
    admin_list =pd.concat(admin_list)
    echech_list=pd.concat(echech_list)
    
    
    return student_enrolmemnt,palmares,admin_list,echech_list
       
```

```python
def radom_student_generator(begin_year=1996,number_year=24):
    student_enrolmemnt = new_enrolment1(began_year=begin_year,end_year=begin_year+number_year,id_levels=[1,2,3,4])
    student_enrolmemnt,palmares,admins,echecs =simulate_promo(student_enrolmemnt,number_year=number_year)
    return student_enrolmemnt,palmares,admins,echecs
```

```python
def total_success(year=2010,id_level=1,vaccation=1):
    conn = create_engine('sqlite:///db.sqlite3', echo=True)
    df = pd.read_sql_table('Passed', conn) 
    df = df[df.year == year+1]
    df = df[df.id_level == id_level
           ]
    df = df[df.vaccation == vaccation]
   
    
    return df.student_id.nunique()




def total_echec(year=2010,id_level=1,vaccation=1):
    conn = create_engine('sqlite:///db.sqlite3', echo=True)
    df = pd.read_sql_table('Failed', conn) 
    df = df[df.year == year+1]
    df = df[df.id_level == id_level+1]
    df = df[df.vaccation == vaccation]
   
    
    return df.student_id.nunique()

def total_enrolment(year=1996,id_level=1,vaccation=1):
    conn = create_engine('sqlite:///db.sqlite3', echo=True)
    df = pd.read_sql_table('Student', conn) 
    df.vaccation=df.vaccation.fillna(0)
    df = df[df.year == year]
    df = df[df.id_level == id_level]
    df = df[df.vaccation == vaccation]
   
    
    return df.student_id.nunique()
```
