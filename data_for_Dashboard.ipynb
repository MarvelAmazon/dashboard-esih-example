```python
def upgrade(df):
    df['passed'] = 0
    for index in range(df.shape[0]):
        if (df.at[index,'grade']>= 65) and (df.at[index,'id_level'] ==1):
            df.at[index,'passed']  = 1
        elif (df.at[index,'grade']>= 70) and (df.at[index,'id_level'] ==2):
            df.at[index,'passed']  = 1
        elif (df.at[index,'grade']>= 75) and (df.at[index,'id_level'] ==3):
            df.at[index,'passed']  = 1
        elif (df.at[index,'grade']>= 80) and (df.at[index,'id_level'] ==4):
            df.at[index,'passed']  = 1
        
    return df
    
```
